```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt     
import seaborn as sns
sns.set_theme(style="whitegrid")
```

```python
ml_dataframe=pd.read_csv('ml_dataframe_clipped_with_new_columns.csv')
```

```python
ml_dataframe.head()
```

```python
ml_dataframe['session_duration_seconds'].describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

print("Preparing the Scatter Plot...")

# STEP 1: Overplotting se bachne aur speed badhane ke liye ek Random Sample lena
# (Agar aapka system fast hai, toh n=100000 bhi kar sakte hain)
sample_df = ml_dataframe.sample(n=50000, random_state=42)

# STEP 2: Graph ki styling aur size set karna
sns.set_theme(style="whitegrid")
plt.figure(figsize=(14, 8))

# STEP 3: Scatter Plot Draw Karna
# Blue (0) Non-buyers ke liye, Red (1) Buyers ke liye
ax = sns.scatterplot(
    data=sample_df,
    x='session_duration_seconds',
    y='total_views',
    hue='target_purchase',
    palette={0: '#3498db', 1: '#e74c3c'}, 
    alpha=0.6,          # Transparency (0 to 1)
    s=50,               # Dot ka size
    edgecolor=None      # Dot ke aas-paas ki line hatana cleaner look ke liye
)

# STEP 4: Labels aur Title
plt.title('User Intent Analysis: Session Duration vs Total Views', fontsize=16, fontweight='bold', pad=15)
plt.xlabel('Session Duration (Seconds)', fontsize=13, fontweight='bold')
plt.ylabel('Total Views in Session', fontsize=13, fontweight='bold')

# Legend (Index) ko customize karna
handles, labels = ax.get_legend_handles_labels()
plt.legend(handles=handles, title='Conversion Status', labels=['Not Purchased (0)', 'Purchased (1)'], loc='upper right', fontsize=11)

plt.tight_layout()
plt.show()
print("Graph Plotted Successfully! 📊")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

print("Preparing the Enhanced Scatter Plot...")

# Random Sample lena (50k ya 1 Lakh)
sample_df = ml_dataframe.sample(n=50000, random_state=42)

# MAGIC STEP: Data ko sort karna taaki '1' (Purchased) hamesha graph mein sabse UPAR draw ho
sample_df = sample_df.sort_values(by='target_purchase')

sns.set_theme(style="whitegrid")
plt.figure(figsize=(14, 8))

# Scatter Plot Draw Karna
ax = sns.scatterplot(
    data=sample_df,
    x='session_duration_seconds',
    y='total_views',
    hue='target_purchase',
    palette={0: '#a6cbe3', 1: '#ff0000'}, # Blue ko thoda light, Red ko ekdum dark kiya
    alpha=0.85,                           # Opacity badha di
    size='target_purchase',               # Red dots ka size bada karne ke liye
    sizes={0: 30, 1: 70},                 # 0 chote (30), 1 bade (70)
    edgecolor='white',
    linewidth=0.5
)

plt.title('User Intent Analysis: Session Duration vs Total Views', fontsize=16, fontweight='bold', pad=15)
plt.xlabel('Session Duration (Seconds)', fontsize=13, fontweight='bold')
plt.ylabel('Total Views in Session', fontsize=13, fontweight='bold')

handles, labels = ax.get_legend_handles_labels()
# Size legend hata kar sirf color legend rakhna
plt.legend(handles=handles[:2], title='Conversion Status', labels=['Not Purchased (0)', 'Purchased (1)'], loc='upper left', fontsize=11)

plt.tight_layout()
plt.show()
```
